데이터의 분포를 평균 0, 표준편차 1로 변환하는 **Z-score 정규화(표준화)**를 구현한다.

## 이론

$$z = \frac{x - \mu}{\sigma}$$

- $\mu$: 평균 (mean)
- $\sigma$: 표준편차 (std)

Z-score는 각 값이 평균에서 몇 표준편차 떨어져 있는지를 나타낸다.

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50], dtype=float)
mu = data.mean()
sigma = data.std()

print(f'평균: {mu}, 표준편차: {sigma}')
```

```python
def z_score(x: np.ndarray) -> np.ndarray:
    return (x - x.mean()) / x.std()

normalized = z_score(data)
print('원본:', data)
print('정규화:', normalized)
print('평균 확인:', normalized.mean().round(10))
print('표준편차 확인:', normalized.std().round(10))
```

## 시각화

`matplotlib`으로 원본과 정규화된 분포를 비교한다.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].hist(data, bins=5, color='#4a62ff')
axes[0].set_title('원본 데이터')
axes[1].hist(normalized, bins=5, color='#dc00c9')
axes[1].set_title('Z-score 정규화')
plt.tight_layout()
plt.show()
```

## 실전 적용: 여러 특성 동시 정규화

데이터셋의 각 특성(열)을 독립적으로 정규화한다.

```python
# 키(cm), 몸무게(kg), 나이 가상 데이터
X = np.array([
    [170, 65, 25],
    [180, 80, 30],
    [160, 55, 22],
    [175, 70, 28],
], dtype=float)

# axis=0: 각 열(특성)별로 정규화
X_norm = (X - X.mean(axis=0)) / X.std(axis=0)

print('정규화 결과:')
print(X_norm.round(3))
print('열별 평균:', X_norm.mean(axis=0).round(10))
print('열별 표준편차:', X_norm.std(axis=0).round(10))
```
